```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from keras.layers import Dropout
import operator
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
no_of_words = 9000
```

```python
data = pd.read_csv('spam.csv' , encoding='latin-1')
data = np.array(data)
x = data[:,1:2]
y = data[:,0:1]

for i in range(len(y)):
    if y[i][0] == 'ham' :
        y[i][0] = 0
    else:
        y[i][0] = 1


para = ''
for i in range(len(x)):
    para += x[i][0].strip() + ' '

para = para.replace('.','').replace('?','').replace("'",'').replace(',','').replace('-','').replace('+','').replace('!','')
para = para.replace('*','').replace('/','').replace('~','').replace(':','').replace(')','').replace('()','').replace('#','')

all_worlds = para.lower().split(' ')
dictionary = {}
for item in all_worlds:
    if item in dictionary:
        dictionary[item] += 1
    else:
        dictionary[item] = 1
dictionary
sorted_dictionary = sorted(dictionary.items(), key=operator.itemgetter(1))
sorted_dictionary = sorted_dictionary[::-1]

tempList = []
for i in range(len(sorted_dictionary)):
    if len(sorted_dictionary[i][0]) >= 1 and "&" not in sorted_dictionary[i][0]:
        tempList.append(sorted_dictionary[i])
tempList = tempList[0:no_of_words]
len(tempList)

final_input = np.atleast_2d()
for i in range(len(x)) :
    tempData = []
    str =x[i][0].replace('.','').replace('?','').replace("'",'').replace(',','').replace('-','').replace('+','').replace('!','').replace('&','')
    str = str.replace('*','').replace('/','').replace('~','').replace(':','').replace(')','').replace('()','').replace('#','')
    for j in range(len(tempList)) :
        if tempList[j][0] not in str :
            tempData.append(0)
        else :
            tempData.append(1)
    final_input.append(tempData)
    
final_input = np.array(final_input)
final_input.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(final_input,y,test_size=0.1 , random_state=0)
x_temp_test = x_test
y_temp_test = y_test
y_train.shape
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
classifier = Sequential()

classifier.add(Dense(units=50,activation='relu',kernel_initializer='uniform',input_dim=no_of_words))
classifier.add(Dropout(rate=0.5))
classifier.add(Dense(units=25,activation='relu',kernel_initializer='uniform'))
classifier.add(Dropout(rate=0.5))
classifier.add(Dense(units=10,activation='relu',kernel_initializer='uniform'))
classifier.add(Dropout(rate=0.5))
classifier.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))

#opt = optimizers.SGD(lr=0.01)
classifier.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')

classifier.fit(x_train,y_train,batch_size=128,epochs=100)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
model_json = classifier.to_json()
with open("spam_model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
classifier.save_weights("spam_model.h5")
print("Saved model to disk")
```
